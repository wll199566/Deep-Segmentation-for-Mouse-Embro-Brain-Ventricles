```python
import os
import numpy as np
from scipy import ndimage
from skimage import io
from matplotlib import pyplot as plt
import pickle
import nibabel as nib
from scipy.ndimage.interpolation import zoom
import random
```

```python
def pickle_store(file_name,save_data):
    fileObj = open(os.path.join(os.getcwd(),'data',file_name),'wb')
    pickle.dump(save_data,fileObj)
    fileObj.close()

def even_func(x):
    if x % 2 == 0:
        return x
    else:
        return x+1

def padded_minimum_size(img, filtered_img, label, min_size):
    img_size= np.shape(img)
    x_offset=0
    y_offset=0
    z_offset=0
    x_flag=False
    y_flag=False
    z_flag=False
    if img_size[0]<min_size:
        x_offset=min_size-img_size[0]
        x_flag=True
    if img_size[1]<min_size:
        y_offset=min_size-img_size[1]
        y_flag=True
    if img_size[2]<min_size:
        z_offset=min_size-img_size[2]
        z_flag=True
    
    if x_flag == True and y_flag == True and z_flag == True:
        padded_img=np.zeros((min_size,min_size,min_size),np.float32)
        padded_filtered_img=np.zeros((min_size,min_size,min_size),np.float32)
        padded_label=np.zeros((min_size,min_size,min_size),np.uint8)
        
        padded_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                   round(y_offset/2):round(y_offset/2)+img_size[1],
                   round(z_offset/2):round(z_offset/2)+img_size[2]]=img
        
        padded_filtered_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                            round(y_offset/2):round(y_offset/2)+img_size[1],
                            round(z_offset/2):round(z_offset/2)+img_size[2]]=filtered_img
        
        padded_label[round(x_offset/2):round(x_offset/2)+img_size[0], 
                     round(y_offset/2):round(y_offset/2)+img_size[1], 
                     round(z_offset/2):round(z_offset/2)+img_size[2]]=label
        
        return padded_img, padded_filtered_img, padded_label
    
    elif x_flag == True and y_flag == True:
        padded_img=np.zeros((min_size,min_size,even_func(img_size[2])),np.float32)
        padded_filtered_img=np.zeros((min_size,min_size,even_func(img_size[2])),np.float32)
        padded_label=np.zeros((min_size,min_size,even_func(img_size[2])),np.uint8)
        
        padded_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                   round(y_offset/2):round(y_offset/2)+img_size[1],
                   0:img_size[2]] = img
        
        padded_filtered_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                            round(y_offset/2):round(y_offset/2)+img_size[1],
                            0:img_size[2]] = filtered_img
        
        padded_label[round(x_offset/2):round(x_offset/2)+img_size[0],
                     round(y_offset/2):round(y_offset/2)+img_size[1],
                     0:img_size[2]] = label
        return padded_img, padded_filtered_img, padded_label
    
    elif x_flag == True and z_flag == True:
        padded_img=np.zeros((min_size,even_func(img_size[1]),min_size),np.float32)
        padded_filtered_img=np.zeros((min_size,even_func(img_size[1]),min_size),np.float32)
        padded_label=np.zeros((min_size,even_func(img_size[1]),min_size),np.uint8)
        
        padded_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                   0:img_size[1],
                   round(z_offset/2):round(z_offset/2)+img_size[2]]=img
        
        padded_filtered_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                            0:img_size[1],
                            round(z_offset/2):round(z_offset/2)+img_size[2]]=filtered_img
        
        padded_label[round(x_offset/2):round(x_offset/2)+img_size[0],
                     0:img_size[1],
                     round(z_offset/2):round(z_offset/2)+img_size[2]]=label
        return padded_img, padded_filtered_img, padded_label
    
    elif y_flag == True and z_flag == True:
        padded_img=np.zeros((even_func(img_size[0]),min_size,min_size),np.float32)
        padded_filtered_img=np.zeros((even_func(img_size[0]),min_size,min_size),np.float32)
        padded_label=np.zeros((even_func(img_size[0]),min_size,min_size),np.uint8)
        
        padded_img[0:img_size[0],
                   round(y_offset/2):round(y_offset/2)+img_size[1],
                   round(z_offset/2):round(z_offset/2)+img_size[2]]=img
        
        padded_filtered_img[0:img_size[0],
                            round(y_offset/2):round(y_offset/2)+img_size[1],
                            round(z_offset/2):round(z_offset/2)+img_size[2]]=filtered_img
        
        padded_label[0:img_size[0],
                     round(y_offset/2):round(y_offset/2)+img_size[1],
                     round(z_offset/2):round(z_offset/2)+img_size[2]]=label
        return padded_img, padded_filtered_img, padded_label
    
    elif x_flag == True:
        padded_img=np.zeros((min_size,even_fuc(img_size[1]),even_func(img_size[2])),np.float32)
        padded_filtered_img=np.zeros((min_size,even_fuc(img_size[1]),even_func(img_size[2])),np.float32)
        padded_label=np.zeros((min_size,even_fuc(img_size[1]),even_func(img_size[2])),np.uint8)
        padded_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                   0:img_size[1],
                   0:img_size[2]]=img
        padded_filtered_img[round(x_offset/2):round(x_offset/2)+img_size[0],
                            0:img_size[1],
                            0:img_size[2]]=filtered_img
        padded_label[round(x_offset/2):round(x_offset/2)+img_size[0],
                     0:img_size[1],
                     0:img_size[2]]=label
        return padded_img, padded_filtered_img, padded_label
    
    elif y_flag == True:
        padded_img=np.zeros((even_func(img_size[0]),min_size,even_func(img_size[2])),np.float32)
        padded_filtered_img=np.zeros((even_func(img_size[0]),min_size,even_func(img_size[2])),np.float32)
        padded_label=np.zeros((even_func(img_size[0]),min_size,even_func(img_size[2])),np.uint8)
        padded_img[0:img_size[0],
                   round(y_offset/2):round(y_offset/2)+img_size[1],
                   0:img_size[2]]=img
        padded_filtered_img[0:img_size[0],
                            round(y_offset/2):round(y_offset/2)+img_size[1],
                            0:img_size[2]]=filtered_img
        padded_label[0:img_size[0],
                     round(y_offset/2):round(y_offset/2)+img_size[1],
                     0:img_size[2]]=label
        return padded_img, padded_filtered_img, padded_label
    
    elif z_flag == True:
        padded_img=np.zeros((even_func(img_size[0]),even_func(img_size[1]),min_size),np.float32)
        padded_filtered_img=np.zeros((even_func(img_size[0]),even_func(img_size[1]),min_size),np.float32)
        padded_label=np.zeros((even_func(img_size[0]),even_func(img_size[1]),min_size),np.uint8)
        padded_img[0:img_size[0],
                   0:img_size[1],
                   round(z_offset/2):round(z_offset/2)+img_size[2]]=img
        padded_filtered_img[0:img_size[0],
                            0:img_size[1],
                            round(z_offset/2):round(z_offset/2)+img_size[2]]=filtered_img
        padded_label[0:img_size[0],
                     0:img_size[1],
                     round(z_offset/2):round(z_offset/2)+img_size[2]]=label
        return padded_img, padded_filtered_img, padded_label
    
    else:
        padded_img=np.zeros((even_func(img_size[0]),even_func(img_size[1]),even_func(img_size[2])),np.float32)
        padded_filtered_img=np.zeros((even_func(img_size[0]),even_func(img_size[1]),even_func(img_size[2])),np.float32)
        padded_label=np.zeros((even_func(img_size[0]),even_func(img_size[1]),even_func(img_size[2])),np.uint8)
        padded_img[0:img_size[0],
                   0:img_size[1],
                   0:img_size[2]]=img
        padded_filtered_img[0:img_size[0],
                            0:img_size[1],
                            0:img_size[2]]=filtered_img
        padded_label[0:img_size[0],
                     0:img_size[1],
                     0:img_size[2]]=label
        return padded_img, padded_filtered_img, padded_label

def extract_positive(whole_img,whole_label,box_size,smallest_ratio=0.99,step_size=2,verbal=False):
    img_size=np.shape(whole_img)
    bv_voxel_num=np.sum(whole_label)
    positive_sub_volumes=[]
    x_slice,y_slice,z_slice = ndimage.find_objects(whole_label)[0]
    offset=5
    
    x_start = x_slice.stop-offset if x_slice.stop-offset >= box_size else box_size
    x_stop = x_slice.start+box_size+offset if x_slice.start+box_size+offset <= img_size[0] else img_size[0]
    
    y_start = y_slice.stop-offset if y_slice.stop-offset >= box_size else box_size
    y_stop = y_slice.start+box_size+offset if y_slice.start+box_size+offset <= img_size[1] else img_size[1]
    
    z_start = z_slice.stop-offset if z_slice.stop-offset >= box_size else box_size
    z_stop = z_slice.start+box_size+offset if z_slice.start+box_size+offset <= img_size[2] else img_size[2]
    
    for i in range(x_start,x_stop+1,step_size):
        for j in range(y_start,y_stop+1,step_size):
            for k in range(z_start,z_stop+1,step_size):
                contain_ratio = np.sum(whole_label[i-box_size:i,
                                                   j-box_size:j,
                                                   k-box_size:k])/(bv_voxel_num+0.001)
                if verbal==True:
                    print('Current box contain bv ratio: ',contain_ratio)
                if contain_ratio > smallest_ratio:
                    positive_sub_volumes.append((i,j,k))
    return positive_sub_volumes


def extract_negetive(whole_img,whole_label,box_size,largest_ratio=0.80,step_size=3,extract_prob=0.80,verbal=False):
    img_size=np.shape(whole_img)
    bv_voxel_num=np.sum(whole_label)
    negative_sub_volumes=[]
    
    x_start = box_size
    x_stop = img_size[0]
    
    y_start = box_size
    y_stop = img_size[1]
    
    z_start = box_size
    z_stop = img_size[2]
    
    for i in range(x_start,x_stop+1,step_size):
        for j in range(y_start,y_stop+1,step_size):
            for k in range(z_start,z_stop+1,step_size):
                contain_ratio = np.sum(whole_label[i-box_size:i,
                                                   j-box_size:j,
                                                   k-box_size:k])/(bv_voxel_num+0.001)
                if verbal==True:
                    print('Current box contain bv ratio: ',contain_ratio)
                if (contain_ratio < largest_ratio and random.uniform(0,1)<extract_prob):
                    negative_sub_volumes.append((i,j,k))
    return negative_sub_volumes
```
